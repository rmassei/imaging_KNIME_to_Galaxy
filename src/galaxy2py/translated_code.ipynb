````
Below is a Python Jupyter notebook that executes the same sequence of analysis steps as the provided Galaxy workflow.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters, exposure, measure, feature
from skimage.color import gray2rgb
from skimage.util import img_as_ubyte
from skimage.morphology import disk, closing

# Load the input image
input_image_path = "path_to_your_input_image.tif"  # Replace with your image path
input_image = io.imread(input_image_path, plugin="tifffile")

# Convert the image to grayscale and extract the first channel
image_gray = input_image[:, :, 0]

# Apply Gaussian filter
image_filtered = filters.gaussian(image_gray, sigma=3.0)

# Apply Contrast Limited Adaptive Histogram Equalization (CLAHE)
image_equalized = exposure.equalize_adapthist(image_filtered, clip_limit=0.03)

# Apply Otsu thresholding
threshold = filters.threshold_otsu(image_equalized)
image_thresholded = image_equalized > threshold

# Convert binary image to label map
label_map = measure.label(image_thresholded)

# Extract image features
features = measure.regionprops(label_map)

# Print the features
for feature in features:
    print(f"Label: {feature.label}, Area: {feature.area}, Perimeter: {feature.perimeter}, Eccentricity: {feature.eccentricity}, Major Axis Length: {feature.major_axis_length}, Minor Axis Length: {feature.minor_axis_length}")

# Display the original image, filtered image, equalized image, thresholded image, and label map
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
axes[0, 0].imshow(input_image)
axes[0, 0].set_title("Original Image")
axes[0, 1].imshow(image_filtered, cmap="gray")
axes[0, 1].set_title("Filtered Image")
axes[0, 2].imshow(image_equalized, cmap="gray")
axes[0, 2].set_title("Equalized Image")
axes[1, 0].imshow(image_thresholded, cmap="gray")
axes[1, 0].set_title("Thresholded Image")
axes[1, 1].imshow(label_map, cmap="gray")
axes[1, 1].set_title("Label Map")
axes[1, 2].axis("off")
plt.show()
```

This notebook performs the following steps:

1.  Loads the input image using `skimage.io.imread`.
2.  Converts the image to grayscale and extracts the first channel.
3.  Applies a Gaussian filter to the grayscale image using `skimage.filters.gaussian`.
4.  Applies Contrast Limited Adaptive Histogram Equalization (CLAHE) to the filtered image using `skimage.exposure.equalize_adapthist`.
5.  Applies Otsu thresholding to the equalized image using `skimage.filters.threshold_otsu`.
6.  Converts the binary thresholded image to a label map using `skimage.measure.label`.
7.  Extracts image features from the label map using `skimage.measure.regionprops`.
8.  Prints the extracted features.
9.  Displays the original image, filtered image, equalized image, thresholded image, and label map using `matplotlib.pyplot`.

Make sure to replace `"path_to_your_input_image.tif"` with the actual path to your input image. This notebook assumes that the input image is in TIFF format. If your image is in a different format, you may need to adjust the `io.imread` function accordingly.
````
