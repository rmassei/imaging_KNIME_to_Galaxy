````python
def prompt_github_models(message:str, model="gpt-4.1"):
    """A prompt helper function that sends a message to ollama and returns only the text response."""
    import os
    import openai
    if isinstance(message, str):
        message = [{"role": "user", "content": message}]
        
    # setup connection to the LLM
    client = openai.OpenAI(
        base_url = "https://models.inference.ai.azure.com/",
        api_key = os.environ.get("GITHUB_MODELS_API_KEY") 
       
    )
    response = client.chat.completions.create(
        model=model,
        messages=message
    )
    
    # extract answer
    return response.choices[0].message.content
````

````python
def load_file(file_path):
    """
    Load a file and return its content.
    
    :param file_path: Path to the file to be loaded.
    :return: Content of the file.
    """
    with open(file_path, 'r') as file:
        content = file.read()
    return content
````

````python
prompt = f"""
You are an excellent python programmer who knows how Glaxy works. You are about to translate a Galaxy workflow into a Python script.
# Example 1:
galaxy workflow:
{load_file('galaxy1.txt')}
summary of the workflow:
1. Name of the workflow: Feature extraction from labeled image
2. What are the input variables and their types?
   - input_label: tiff
   - input_raw: tiff
3. What are the output variables and their types?
    - measurement_table: csv
    
corresponding Python script:
{load_file('python1.txt')}

# Example 2:
galaxy workflow:
{load_file('galaxy2.txt')}
summary of the workflow:
1. Name of the workflow: Convert binary image to label map
2. What are the input variables and their types?
   - binary_image: tiff
3. What are the output variables and their types?
   - label_image: tiff

corresponding Python script:
{load_file('python2.txt')}   

# Your task:
{load_file('galaxy3.txt')}

First, summarize the Galaxy workflow in bullet points. Specifiy input and output variables and their types. 

Example summary:
1. Name of the workflow:
2. What are the input variables and their types?
3. What are the output variables and their types?

Then, after two new lines, write the corresponding Python script that implements the same functionality.
"""
````

````python
print(prompt_github_models(prompt))
````

````python
import re

def extract_code_blocks(markdown_text):
    # Regulärer Ausdruck für dreifache Backticks mit optionalem Sprachenhinweis
    pattern = r"```(?:\w+)?\n(.*?)```"
    code_blocks = re.findall(pattern, markdown_text, re.DOTALL)
    return code_blocks[-1]

# Beispiel
markdown = """
Hier ist ein bisschen Text.

```python
def hello():
    print("Hello, world!")
```
"""
````

````python
print(extract_code_blocks(prompt_github_models(prompt)))
````

````python
extract_code_blocks(markdown)
````
